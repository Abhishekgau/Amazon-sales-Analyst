# Amazon sales data analyst project

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("Superstore.csv",encoding="latin-1")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

# converting order Date columns

```python
data["Order Date"] = pd.to_datetime(data["Order Date"])
```

```python
data.info()
```

# converting ship date columns

```python
data["Ship Date"] = pd.to_datetime(data["Ship Date"])
```

```python
data.info()
```

```python
 data.head()
```

# creating three columns name is monthly and year and order date of weeks 

```python
data['order Month'] = data['Order Date'].dt.month
data['order year'] = data['Order Date'].dt.year
data['Order Day of weeks'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

### you need to calculate the monthly sales of the store and identify which month had the highest sales and which month had the lowest sales

```python
sales_by_month = data.groupby('order Month')['Sales'].sum().reset_index()
```

```python
sales_by_month
```

```python
fig = px.line(sales_by_month,x = 'order Month',y = 'Sales',title = 'Monthly Sales Analysis')
fig.show()
```

### you need to analyze sales based on product categoties and determine which category has the lowest sales and which category has the highest sales

```python
sales_of_category   = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_of_category
```

```python
fig = px.pie(sales_of_category,values = "Sales", names = "Category",hole = 0.1,color_discrete_sequence = px.colors.qualitative.Pastel)
```

```python
fig.update_traces(textposition = 'inside',textinfo = 'percent+label')
fig.update_layout(title_text = 'sales Analysis by category', title_font= dict(size=24))
fig.show()
```

### The sales analysis needs to be done based on sub-categories

```python
data.info()
```

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory, x= 'Sub-Category',y = 'Sales',title = 'Sales by sub-category')
fig.show()
```

### you need to analyze the monthly profit from sales and determine which month had the highest profit

```python
profit_by_month = data.groupby('order Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
fig = px.bar(profit_by_month,x = 'order Month',y = 'Profit',title  ='Profit by Month sales')
fig.show()
```

### Analyze the sales and profit by customer segment

```python
data.head()
```

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()

```

```python
sales_profit_by_segment
```

```python

color_palette = px.colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment',
                  yaxis_title='Amount')
fig.show()
```

### Analyze the sale  to profit ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales_to_Profit_Ratio']])
```
